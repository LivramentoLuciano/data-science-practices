# Análisis del "Registro de nacimientos en Estados Unidos (1910-2020)"

## Carga y Previsualización de los datos

```python
# librerias a utilizar
import pandas as pd
from google.cloud import bigquery
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# cliente bigquery para consultar database
client = bigquery.Client()

# Cargo la tabla de nombres que me interesa
dataset_ref = client.dataset('usa_names', project='bigquery-public-data')
table_ref = dataset_ref.table('usa_1910_current')
names_table = client.get_table(table_ref)
                               
# Cabecera de la tabla, para inspeccionar datos
client.list_rows(names_table, max_results=5).to_dataframe()
```

## Principales características. Demografía.
En primer lugar, se realiza un breve resumen de la información disponible, observando: cuántas inscripciones se han realizado en Estados Unidos de 1910 a la actualidad (2020), proporción según sexo y según Estado.

```python
# Evito superar cuota de lectura de BigQuery
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)
```

### Histórico de personas registradas
Se obtiene el total histórico de personas registradas y la proporción por género. Además, se observa la evolución año a año de estos valores.

```python
# Registros nacionales, año a año, por genero
# nota: lo habia hecho con subquery pero es mas lento y largo
hist_people_query = '''
SELECT year, gender, SUM(number) AS year_registers
FROM `bigquery-public-data.usa_names.usa_1910_current`
GROUP BY year, gender
ORDER BY year ASC
'''

hist_people_query_job = client.query(hist_people_query, job_config=safe_config)
hist_people = hist_people_query_job.to_dataframe()

# distribuyo en columnas por género y agrego los totales (para graficar)
hist_people_full = hist_people.pivot('year', 'gender', 'year_registers')
hist_people_full['ALL'] = hist_people_full['F'] +hist_people_full['M']
hist_people_full.head(3)
```

#### Registros totales acumulados (1910-2020) y proporción por género

```python
# Registros totales en toda la historia
total_regist = hist_people_full['ALL'].sum()
fem_regist_perc = hist_people_full['F'].sum() / total_regist * 100
male_regist_perc = hist_people_full['M'].sum() / total_regist * 100

# print('En Estados Unidos se han registrado un total de {:,} personas desde 1910 a la actualidad.'.format(total_regist)) 
# print('Siendo el {:.1f}% mujeres y el restante {:.1f}% hombres'.format(fem_regist_perc, male_regist_perc))
hist_overview = pd.DataFrame(data={'Female [%]': fem_regist_perc, 'Male [%]': male_regist_perc, 'All': total_regist}, index=['Acumulado Histórico'])
hist_overview
```

### Histórico de personas registradas por Estado.
Analizo la demografía del país, por Estado. Se desea obtener así, información sobre los estados donde más personas nacen año a año y proporciones en cuanto al género.

```python
# Registros estaduales, año a año, por genero
hist_people_states_query = '''
SELECT year, state, SUM(number) AS registers
FROM `bigquery-public-data.usa_names.usa_1910_current`
GROUP BY year, state
ORDER BY year ASC
'''

hist_people_states_query_job = client.query(hist_people_states_query, job_config=safe_config)
hist_people_states = hist_people_states_query_job.to_dataframe()

# distribuyo en columnas por estado
hist_people_states = hist_people_states.pivot('year', 'state', 'registers')
hist_people_states.head(3)
```

#### Mapa de calor. Estados de mayor cantidad de nacimientos a lo largo del tiempo.
Se visualiza, a continuación, el progreso de la cantidad de nacimientos a lo largo del tiempo, en todos los Estados. Esto permite, por ejemplo, observar rápidamente, cuáles son los Estados con una mayor cantidad de nacimientos en cada época.

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=hist_people_states)
plt.title('Cantidad de nacimientos por año y Estado')
plt.xlabel('Estado')
plt.ylabel('Nacimientos')
```

#### Nacimientos registrados por año y Estado, según género.
Al análisis anterior, ahora se incorpora la distinción por género

Para no tener que realizar 2 gráficos, uno para hombres y otro para mújeres, se realiza el cálculo de una "relación male/fem" (hombres/mujeres), de tal manera que, cuanto mayor sea "male/fem", mayor será la cantidad de hombres sobre la de mujeres y viceversa. 

```python
# Nacimientos agrupados por Año y Estado
# Genero columnas con los totales por género y el ratio 'male_fem'
# La mejor forma de hacerlo que encontre fue con subquerys
hist_births_states_gender_query = '''
WITH female_state_registers AS (
    SELECT year, state, SUM(number) AS f_registers
    FROM `bigquery-public-data.usa_names.usa_1910_current`
    WHERE gender = 'F'
    GROUP BY year, state
    ORDER BY year ASC
), 
male_state_registers AS (
    SELECT year, state, SUM(number) AS m_registers
    FROM `bigquery-public-data.usa_names.usa_1910_current`
    WHERE gender = 'M'
    GROUP BY year, state
    ORDER BY year ASC
)

SELECT msr.year,
       msr.state, 
       f_registers,
       m_registers,
       (f_registers + m_registers) AS total_registers,
       (m_Registers / f_registers) AS male_fem_ratio
FROM male_state_registers AS msr
FULL JOIN female_state_registers AS fsr 
ON msr.year = fsr.year AND msr.state = fsr.state
ORDER BY msr.year ASC, state ASC
'''

hist_births_states_gender_query_job = client.query(hist_births_states_gender_query, job_config=safe_config)
hist_births_states_gender = hist_births_states_gender_query_job.to_dataframe()
hist_births_states_gender.head(3)
```

#### Gráfico

```python
# Para poder graficarlo (Heatmap), es necesario primero pivotearlo
# año vs state, con 'male_fem_ratio' como valor
hist_births_states_gender_ratio = hist_births_states_gender.pivot('year', 'state', 'male_fem_ratio')

plt.figure(figsize=(12,6))
sns.heatmap(data=hist_births_states_gender_ratio, vmin=0.5, vmax=1.3)
plt.title('Ratio hombres/mujeres nacidxs por año y Estado')
plt.xlabel('Estado')
plt.ylabel('Ratio h/m')
```
